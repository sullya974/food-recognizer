```python
import tensorflow.keras.backend as K
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

## Predicting classes from new images from the internet using the best train model

```python
# Load the best save model to make prediction
K.clear_session()
model_best = load_model("best_model_58class_all.hdf5", compile=False)
```

```python
def predict_class(model, images, show=True):
    for img in images:
        print(f"Image : {img}")
        img = image.load_img(img, target_size=(299, 299))
        img = image.img_to_array(img)
        img = np.expand_dims(img, axis=0)
        img /= 255
        
        pred = model.predict(img) # Le modèle lance sa prédiction
        index = np.argmax(pred)   # Récupère l'indice de la valeur la plus élévée de la prediction
        print(f"Index : {index}")
        food_list.sort()          # Tri la liste de food de façon croissante
        pred_value = food_list[index]             # Récupère l'élément corespondant à l'index à la probabilité la plus élevé
                                  # dans la liste de food             
        if show:
            plt.imshow(img[0])
            plt.axis("off")
            plt.title(pred_value)
            plt.show()
```

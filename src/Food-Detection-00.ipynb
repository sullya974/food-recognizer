<h2> Classification avec LeNet5

```python
def load_element_utils(file_path):
    with open(file_path, "r") as f:
        result = [line.strip() for line in f]
    return result
```

```python
classes = load_element_utils("../datasets/food41/meta/meta/classes.txt")
classes
```

```python
dict_classes_idx = {}
for idx, val in enumerate(classes):
    dict_classes_idx[val] = idx
#     print(f"{idx} : {val}")
    
dict_classes_idx    
```

```python
labels = load_element_utils("../datasets/food41/meta/meta/labels.txt")
labels
```

```python
images_path_root = "../datasets/food41/images"
train_list = [f"{images_path_root}/{img_path}.jpg" for img_path in load_element_utils("../datasets/food41/meta/meta/train.txt")]
train_list
```

```python
test_list = [f"{images_path_root}/{img_path}.jpg" for img_path in load_element_utils("../datasets/food41/meta/meta/test.txt")]
test_list
```

```python
def get_X(dataset, X_length):
    X = [plt.imread(img_path) for img_path in dataset[:X_length]]
    return X
```

```python
def get_y(dataset, y_length):
    y = [dict_classes_idx[img_path.split("/")[4]] for img_path in dataset[:y_length]]
    return y
```

```python
X_train = get_X(train_list, 2000)
X_train
```

```python
y_train = get_y(train_list, 2000)
```

```python
# X_test = [plt.imread(img_path) for img_path in test_list[:1000]]
X_test = get_X(test_list, 1000)
X_test
```

```python
y_test = get_y(test_list, 1000)
y_test
```
